<h3>Boxplot</h3>

inf_provincia

```python
acc_vel = ['HASTA 512 kbps', '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps', '+ 6 Mbps - 10 Mbps', '+ 10 Mbps - 20 Mbps', '+ 20 Mbps - 30 Mbps', '+ 30 Mbps',
       'Otras Velocidades', 'Mbps (Media de bajada)']

inf_provincia[acc_vel].boxplot(figsize=(12, 6))
plt.title('Boxplot de Velocidades de Conexión a Nivel Provincial')
plt.ylabel('Velocidad (Mbps)')
plt.xticks(rotation=45)
plt.show()


acc_tel = ['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros']

inf_provincia[acc_tel].boxplot(figsize=(12, 6))
plt.title('Boxplot de Tipos de Acceso a Internet a Nivel Nacional')
plt.ylabel('Cantidad de Conexiones')
plt.xticks(rotation=45)
plt.show()

Dial_BAf = ['Banda ancha fija', 'Dial up']

inf_provincia[Dial_BAf].boxplot(figsize=(12, 6))
plt.title('Boxplot de Tipos de Acceso BA-Dial a Nivel Provincial')
plt.ylabel('Conexiones')
plt.xticks(rotation=45)
plt.show()


Accesos = ['Accesos por cada 100 hogares', 'Accesos por cada 100 hab']
inf_provincia[Accesos].boxplot(figsize=(12, 6))
plt.title('Boxplot de Acceso a Internet a Nivel Provincial')
plt.ylabel('Conexiones')
plt.xticks(rotation=45)
plt.show()
```

```python
En los gráficos de VígoTs presentados, se observan puntos inusualmente elevados en la parte superior de algunas visualizaciones.
```

<h2>Tendencias en el tiempo:</h2>

suma_informacion_nacional

```python
columnas_velocidad = ['Año', 'HASTA 512 kbps', '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps', '+ 6 Mbps - 10 Mbps', '+ 10 Mbps - 20 Mbps', '+ 20 Mbps - 30 Mbps', 
                      '+ 30 Mbps', 'Otras Velocidades', 'Total Velocidades', 'Mbps (Media de bajada)']

df_velocidad = inf_provincia[columnas_velocidad]

df_velocidad_mean = df_velocidad.groupby('Año').mean()

df_velocidad_mean.plot(figsize=(15, 10))
plt.title('Tendencias de Velocidades a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Media de velocidades (Mbps)')
plt.legend(title='Velocidades')
plt.show()
```

```python
columnas_acceso = [ 'Año', 'ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros', 'Total Accesos']

df_velocidad = inf_provincia[columnas_acceso]

df_velocidad_mean = df_velocidad.groupby('Año').mean()

df_velocidad_mean.plot(figsize=(15, 10))
plt.title('Tendencias de Acceso a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Media de velocidades (Mbps)')
plt.legend(title='Velocidades')
plt.show()
```

```python
columnas_acceso_DialBA = ['Año', 'Banda ancha fija', 'Dial up', 'Total dial_BAf']

df_velocidad = inf_provincia[columnas_acceso_DialBA]

df_velocidad_mean = df_velocidad.groupby('Año').mean()

df_velocidad_mean.plot(figsize=(15, 10))
plt.title('Tendencias de Velocidades a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Media de velocidades (Mbps)')
plt.legend(title='Velocidades')
plt.show()
```

```python
acceso = ['Año', 'Accesos por cada 100 hogares', 'Accesos por cada 100 hab']

df_velocidad = inf_provincia[acceso]

df_velocidad_mean = df_velocidad.groupby('Año').mean()

df_velocidad_mean.plot(figsize=(15, 10))
plt.title('Tendencias de Acceso a la Televisión a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Media de velocidades (Mbps)')
plt.legend(title='Velocidades')
plt.show()
```
